```python
import pygame
import random

# Inicializando o Pygame
pygame.init()

# Definindo as dimensões da tela
largura_tela = 800
altura_tela = 600
tela = pygame.display.set_mode((largura_tela, altura_tela))
pygame.display.set_caption('Pong')

# Cores
preto = (0, 0, 0)
branco = (255, 255, 255)

# Definindo as variáveis para o jogo
paddle_largura = 15
paddle_altura = 100
bola_diametro = 20

# Posições iniciais
paddle_esq_y = altura_tela // 2 - paddle_altura // 2
paddle_dir_y = altura_tela // 2 - paddle_altura // 2
bola_x = largura_tela // 2 - bola_diametro // 2
bola_y = altura_tela // 2 - bola_diametro // 2

# Velocidade
paddle_velocidade = 10
bola_velocidade_x = 5 * random.choice((1, -1))
bola_velocidade_y = 5 * random.choice((1, -1))

# Função para desenhar os paddles e a bola
def desenhar():
    tela.fill(preto)
    pygame.draw.rect(tela, branco, (10, paddle_esq_y, paddle_largura, paddle_altura))  # Paddle esquerdo
    pygame.draw.rect(tela, branco, (largura_tela - 10 - paddle_largura, paddle_dir_y, paddle_largura, paddle_altura))  # Paddle direito
    pygame.draw.ellipse(tela, branco, (bola_x, bola_y, bola_diametro, bola_diametro))  # Bola
    pygame.display.update()

# Loop principal do jogo
rodando = True
while rodando:
    pygame.time.delay(20)  # Controla a taxa de atualização do jogo (FPS)

    # Verificar eventos
    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            rodando = False

    # Movimentação do paddle esquerdo (jogador)
    teclas = pygame.key.get_pressed()
    if teclas[pygame.K_w] and paddle_esq_y > 0:  # Move para cima
        paddle_esq_y -= paddle_velocidade
    if teclas[pygame.K_s] and paddle_esq_y < altura_tela - paddle_altura:  # Move para baixo
        paddle_esq_y += paddle_velocidade

    # Movimentação do paddle direito (computador)
    if bola_y < paddle_dir_y + paddle_altura // 2:
        paddle_dir_y -= paddle_velocidade
    if bola_y > paddle_dir_y + paddle_altura // 2:
        paddle_dir_y += paddle_velocidade

    # Movimentação da bola
    bola_x += bola_velocidade_x
    bola_y += bola_velocidade_y

    # Colisão com as bordas superior e inferior
    if bola_y <= 0 or bola_y + bola_diametro >= altura_tela:
        bola_velocidade_y *= -1

    # Colisão com os paddles
    if bola_x <= 10 + paddle_largura and paddle_esq_y < bola_y + bola_diametro // 2 < paddle_esq_y + paddle_altura:
        bola_velocidade_x *= -1
    if bola_x >= largura_tela - 10 - paddle_largura - bola_diametro and paddle_dir_y < bola_y + bola_diametro // 2 < paddle_dir_y + paddle_altura:
        bola_velocidade_x *= -1

    # Se a bola passar por um paddle (ponto perdido)
    if bola_x <= 0 or bola_x + bola_diametro >= largura_tela:
        bola_x = largura_tela // 2 - bola_diametro // 2
        bola_y = altura_tela // 2 - bola_diametro // 2
        bola_velocidade_x *= random.choice((1, -1))
        bola_velocidade_y *= random.choice((1, -1))

    # Desenhar tudo na tela
    desenhar()

# Finalizar o Pygame
pygame.quit()
```

```python
import pygame
import time
import random

# Inicializando o Pygame
pygame.init()

# Definindo as cores
branco = (255, 255, 255)
preto = (0, 0, 0)
vermelho = (213, 50, 80)
verde = (0, 255, 0)
azul = (50, 153, 213)

# Definindo as dimensões da tela
largura_tela = 600
altura_tela = 400
tela = pygame.display.set_mode((largura_tela, altura_tela))
pygame.display.set_caption('Jogo da Cobrinha')

# Controlando a velocidade do jogo
clock = pygame.time.Clock()
snake_block = 10
snake_speed = 10

# Fonte de texto
font_style = pygame.font.SysFont("bahnschrift", 25)
score_font = pygame.font.SysFont("comicsansms", 35)

# Função para mostrar o texto na tela
def nossa_pontuacao(score):
    value = score_font.render("Pontuação: " + str(score), True, preto)
    tela.blit(value, [0, 0])

# Função para desenhar a cobra
def nossa_cobra(snake_block, snake_lista):
    for x in snake_lista:
        pygame.draw.rect(tela, verde, [x[0], x[1], snake_block, snake_block])

# Função para mostrar mensagens
def nossa_mensagem(msg, cor):
    mesg = font_style.render(msg, True, cor)
    tela.blit(mesg, [largura_tela / 6, altura_tela / 3])

# Função principal do jogo
def gameLoop():
    game_over = False
    game_close = False

    # Posições iniciais da cobra
    x1 = largura_tela / 2
    y1 = altura_tela / 2

    x1_mudanca = 0
    y1_mudanca = 0

    snake_lista = []
    comprimento_da_cobra = 1

    # Gerando a comida
    comida_x = round(random.randrange(0, largura_tela - snake_block) / 10.0) * 10.0
    comida_y = round(random.randrange(0, altura_tela - snake_block) / 10.0) * 10.0

    while not game_over:

        while game_close:
            tela.fill(azul)
            nossa_mensagem("Você perdeu! Pressione Q para sair ou C para jogar novamente", vermelho)
            nossa_pontuacao(comprimento_da_cobra - 1)
            pygame.display.update()

            for evento in pygame.event.get():
                if evento.type == pygame.QUIT:
                    game_over = True
                    game_close = False
                if evento.type == pygame.KEYDOWN:
                    if evento.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if evento.key == pygame.K_c:
                        gameLoop()

        for evento in pygame.event.get():
            if evento.type == pygame.QUIT:
                game_over = True
            if evento.type == pygame.KEYDOWN:
                if evento.key == pygame.K_LEFT:
                    x1_mudanca = -snake_block
                    y1_mudanca = 0
                elif evento.key == pygame.K_RIGHT:
                    x1_mudanca = snake_block
                    y1_mudanca = 0
                elif evento.key == pygame.K_UP:
                    y1_mudanca = -snake_block
                    x1_mudanca = 0
                elif evento.key == pygame.K_DOWN:
                    y1_mudanca = snake_block
                    x1_mudanca = 0

        if x1 >= largura_tela or x1 < 0 or y1 >= altura_tela or y1 < 0:
            game_close = True
        x1 += x1_mudanca
        y1 += y1_mudanca
        tela.fill(azul)
        pygame.draw.rect(tela, vermelho, [comida_x, comida_y, snake_block, snake_block])
        snake_cabeca = []
        snake_cabeca.append(x1)
        snake_cabeca.append(y1)
        snake_lista.append(snake_cabeca)
        if len(snake_lista) > comprimento_da_cobra:
            del snake_lista[0]

        for x in snake_lista[:-1]:
            if x == snake_cabeca:
                game_close = True

        nossa_cobra(snake_block, snake_lista)
        nossa_pontuacao(comprimento_da_cobra - 1)

        pygame.display.update()

        # Checando se a cobra comeu a comida
        if x1 == comida_x and y1 == comida_y:
            comida_x = round(random.randrange(0, largura_tela - snake_block) / 10.0) * 10.0
            comida_y = round(random.randrange(0, altura_tela - snake_block) / 10.0) * 10.0
            comprimento_da_cobra += 1

        clock.tick(snake_speed)

    pygame.quit()
    quit()

# Chamando a função para iniciar o jogo
gameLoop()
```
