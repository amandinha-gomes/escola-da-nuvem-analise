```python
import pandas as pd

df = pd.read_csv('Matriculas_pii_none.csv')

df.head()
```

```python
df['AT - Frequência (%)'] = pd.to_numeric(df['AT - Frequência (%)'], errors='coerce')

bins = [0, 25, 50, 75, 100]
labels = ['0% - 25%', '25% - 50%', '50% - 75%', '75% - 100%']

df['Faixa de Frequência'] = pd.cut(df['AT - Frequência (%)'], bins=bins, labels=labels, right=True, include_lowest=True)

contagem_frequencias = df['Faixa de Frequência'].value_counts().sort_index()

print("Contagem por faixa de frequência:")
print(contagem_frequencias)
```
